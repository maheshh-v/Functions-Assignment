# 1. What is the difference between a function and a method in Python?

In Python, both functions and methods are blocks of reusable code, but there is a key difference in how they are used and where they belong.

- A **function** is a piece of code that is defined independently and can be called on its own. It is not tied to any object.
- A **method** is a function that is associated with an object (especially an object of a class) and is called on that object using the dot (`.`) operator.

### Summary:
- **Function**: Defined using `def` and can be called by name.
- **Method**: Called on an object and often acts on that object’s data.

For example, `len()` is a function, while `"hello".upper()` is a method of the string object.

# 2. Explain the concept of function arguments and parameters in Python.

In Python, **parameters** and **arguments** are terms used in the context of functions, and they are related but different.

- **Parameters** are the variable names listed in the function definition. They act as placeholders.
- **Arguments** are the actual values you pass to the function when you call it.

This distinction is important for understanding how functions receive data.

### Types of Arguments:
- **Positional arguments** – matched by position.
- **Keyword arguments** – matched by name.
- **Default arguments** – have default values if not provided.
- **Variable-length arguments** – `*args` (non-keyword) and `**kwargs` (keyword).

# 3. What are the different ways to define and call a function in Python?

In Python, you define a function using the `def` keyword followed by the function name and parentheses. You can call it by using its name followed by parentheses.

### Ways to define functions:
- **Simple function**: Takes no parameters.
- **Parameterized function**: Takes input parameters.
- **Function with default values**.
- **Variable-length argument function**.
- **Lambda (anonymous) function**.

### Calling a function:
You call a function by writing its name followed by parentheses. If the function requires arguments, you pass them inside the parentheses.

Even though `lambda` is a function, it's mostly used in places where a quick, one-time-use function is required (like with `map()` or `sorted()`).

# 4. What is the purpose of the `return` statement in a Python function?

The `return` statement in Python is used to send a value back to the place where the function was called. When a function reaches a `return` statement, the function execution ends immediately, and the specified value is returned.

### Why is `return` important?
- It makes the result of the function reusable.
- It allows functions to produce an output that can be stored or used in other operations.

If a function doesn’t have a return statement, it returns `None` by default.

---

# 5. What are iterators in Python and how do they differ from iterables?

In Python, an **iterable** is any object capable of returning its members one at a time, like a list or a string. An **iterator** is an object that represents a stream of data; it returns the next item of the sequence when you call `next()` on it.

### Key Differences:
- **Iterable**: Can be passed to the `iter()` function to get an iterator. Examples: lists, tuples, sets.
- **Iterator**: Has the methods `__iter__()` and `__next__()`. Used to iterate through items one at a time.

Once an iterator is exhausted, it cannot be reused unless recreated.

---

# 6. Explain the concept of generators in Python and how they are defined.

A **generator** in Python is a special type of iterator, defined using a function but instead of `return`, it uses the `yield` keyword. This allows it to return values one at a time and pause execution between them.

### Characteristics of Generators:
- They don’t store all values in memory.
- They generate values on the fly and are useful for large datasets.
- Each time `yield` is called, the function’s state is saved.

Generators are defined like regular functions, but use `yield` to return data.

---

# 7. What are the advantages of using generators over regular functions?

Generators have several advantages, especially when dealing with large data or streams of data.

### Advantages:
- **Memory Efficiency**: They don’t store the entire sequence in memory.
- **Lazy Evaluation**: Values are generated only when requested.
- **Improved Performance**: Especially for loops and data processing.
- **Cleaner Code**: Avoids managing iterator state manually.

Generators are ideal for reading large files, producing infinite sequences, or streaming data.

---

# 8. What is a lambda function in Python and when is it typically used?

A **lambda function** is a small, anonymous function defined using the `lambda` keyword. It can take any number of arguments but only has one expression.

### When to use:
- When a small function is needed temporarily.
- Typically used with higher-order functions like `map()`, `filter()`, or `sorted()`.

Lambda functions improve code brevity but can reduce readability if overused.

---

# 9. Explain the purpose and usage of the `map()` function in Python.

The `map()` function is used to apply a function to all items in an iterable (like a list). It returns a new map object (an iterator) with the results.

### Purpose:
- To perform a common operation on each item of an iterable without writing a loop.

**map(function, iterable)** applies the function to each element.

For example, doubling all numbers in a list can be done easily with `map()` instead of using a for-loop.

---

# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

These are **functional programming tools** in Python:

- **`map()`**: Transforms each item by applying a function. Returns the same number of items.
- **`filter()`**: Filters items based on a condition. Returns only items that pass the test.
- **`reduce()`**: Applies a function cumulatively to reduce the iterable to a single value. It must be imported from `functools`.

### Summary:
- `map()` → transforms.
- `filter()` → selects.
- `reduce()` → combines.

These functions are best used when the operation is simple and concise.

```python
# 1. Write a Python function that takes a list of numbers as input
# and returns the sum of all even numbers in the list.

def sum_of_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)

print(sum_of_even([1, 2, 3, 4, 5, 6]))
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(s):
    return s[::-1]

print(reverse_string("hello"))
```

```python
# 3. Implement a Python function that takes a list of integers
# and returns a new list containing the squares of each number.

def square_list(lst):
    return [x**2 for x in lst]

print(square_list([1, 2, 3, 4]))
```

```python
# 4. Write a Python function that checks if numbers from 1 to 200 are prime or not.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for i in range(1, 201):
    if is_prime(i):
        print(f"{i} is Prime")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence
# up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

for num in FibonacciIterator(10):
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(limit):
    for i in range(limit + 1):
        yield 2 ** i

for value in power_of_two(5):
    print(value)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 3), (2, 1), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)  # Output: [(2, 1), (4, 2), (1, 3)]
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (9/5)*c + 32, celsius))
print(fahrenheit)  # Output: [32.0, 68.0, 98.6, 212.0]
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))

print(remove_vowels("Hello World"))  # Output: "Hll Wrld"
```

```python
# 11. Bookshop Accounting: Write a Python program that returns a list with 2-tuples.
# Each tuple consists of the order number and the product of price * quantity.
# Add €10 if the product is less than €100. Use lambda and map.

orders = [
    ["34587", "Learning Python", 4, 40.95],
    ["98762", "Programming Java", 5, 36.95],
    ["77226", "Head First Python", 3, 49.95],
    ["88112", "Intro to Algorithms", 3, 89.95]
]

result = list(map(
    lambda order: (
        order[0],
        order[2]*order[3] if order[2]*order[3] >= 100 else order[2]*order[3] + 10
    ),
    orders
))

print(result)
```
